### zip()函数

```python
# zip()用于将可迭代对象（列表、元祖）作为参数，将对象中对应元素打包成一个个元祖，返回这些元祖组成的列表
# zip() 常与 for 循环搭配使用
# 当可迭代对象是列表
list_a = [1,2,3,4]
list_b = ['a','b','c','d','e']
for i in zip(list_a,list_b):  #有顺序区分
    print(i)
print('------------')
for j in zip(list_b,list_a):
    print(j)
```

```python
# 当可迭代对象是元祖
tup_a = (1,2,3,4)
tup_b = ('a','b','c','d')
for i in zip(tup_a,tup_b):
    print(i)
print('-----------')
for j in zip(tup_b,tup_a):
    print(j)
```

### 算术运算符

```python
# % 返回两数相除的余数
print(10%20)
# ** 返回 x 的 y 次幂
print(10**20)
# //返回两数相除以后商的整数部分
print(10//3)
```

### 比较运算符

```python
# == 等于
# != 不等于
# >= 大于等于
# <= 小于等于
print(10 != 20)
print(10 >= 20)
```

### 逻辑运算符

```python
# and ：a and b，a和b同时为真，结果为真
print((10 != 20)and(10 <= 20))
print((10 != 20)and(10 >= 20))
# or : a or b，a和b只要有一个为真，结果为真
print((10 != 20)or(10 >= 20))
# not : not a，如果a为真，则返回False，否则返回True
print(not (10 != 20))
print(not (10 == 20))
```

## 循环语句

### for 循坏

```python
# for 循环用来遍历任何序列的项目
food = ['烧烤','火锅','锅包肉']
for fo in food:
    print('我想吃{}'.format(fo))
```

### while 循环

```python
# while 循环用来执行程序，当条件满足时，一直执行程序，直到程序不满足，终止程序
day = 1
while day <= 7:
    print('我已经减肥{}天了'.format(day))
    day += 1
print('我已经减肥{}天了，可以去大吃一顿啦'.format(day-1))
```

## 函数

### 普通函数

```python
# 普通函数由函数名（必需）、参数、语句块（必需）、return、变量 组成
def learn_python(location):
    print('我在{}上学python'.format(location))
learn_python('家里')
learn_python('公司')
print('-----------')
print(learn_python('家里'))
```

```python
# 利用return 对语句块的运行结果进行返回
def learn_python(location):
    doing = ('我在{}上学python'.format(location)) #将运行结果赋值给doing
    return doing
print(learn_python('家里'))
print(learn_python('公司'))
```

```python
# 定义含有多个参数的函数
def learn_python(location,person):
    doing = ('我在{}上学python，人{}'.format(location,person)) #将运行结果赋值给doing
    return doing
print(learn_python('家里','就我一个'))
print(learn_python('公司','不多'))
```

### map 函数

```python
# map 函数的表现形式是 map(function,agrs),表示对序列args中的每个值进行function操作，最终得到一个结果序列
a = map(lambda x,y:x+y,[1,2,3],[3,2,1])
print(a)          # map函数生成的结果不会直接显示，想要获取结果需要for 循环遍历
for i in a:
    print(i)
```

## 异常值的检测与处理
### 异常值检测
    根据业务经验划定不同指标的正常范围，超过该范围的值算作异常值：
    1.通过绘制箱型图，把大于、小于箱型图上下边缘的点称为异常值
    2.如果数据服从正态分布：一个数值与平均值之间的偏差超过3倍标准差，那么就是异常值
### 异常值处理
    1.直接删除异常值
    2.异常值当成缺失值填充
    3.异常值当作特殊情况，研究异常值出现的原因

## 数值排名
### Excel实现
    在excel中用于排名的函数有RANK.AVG()\RANK.EQ()两个，两个函数的不同在于处理重复值方式不同
    RANK.AVG(number,ref,order)：number表示待排名的数值，ref表示一整列数值的范围，order表示指明降序还是升序，当排名有重复值时返回重复值的平均排名；
    RANK.EQ(number,ref,order)：number表示待排名的数值，ref表示一整列数值的范围，order表示指明降序还是升序，当排名有重复值时返回重复值的最佳排名
### python实现
    rank(method,ascending):ascending用来指明升序或是降序，默认升序；
    merhod可取的不同参数值说明：
    average：与Excel中的RANK.AVG函数功能一样
    first：按值在所有的待排列数据中出现的先后顺序排名
    min：与Excel中的RANK.EQ函数功能一样
    max：与min相反，取重复值的最大排名

```python
import pandas as pd
df = pd.read_csv(r'/Users/mc-016/Desktop/对比Excel,轻松学习Python数据分析数据集/MappingAnalysis_Data.csv',
                encoding = 'gbk')
df
```

```python
# method取值average
print(df['销售ID'].rank(method = 'average'))
df['销售ID'].rank(method = 'average',ascending = False)
```

```python
# method取值first
df['销售ID'].rank(method = 'first')
```

```python
# method取值min
df['销售ID'].rank(method = 'min')
```

```python
# method取值max
df['销售ID'].rank(method = 'max')
```
